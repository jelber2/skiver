```python
import pandas as pd

kvmer_df = pd.read_csv("../ERR3152366.csv")
kvmer_ref_df = pd.read_csv("../ERR3152366_ref.csv")
```

```python
kvmer_df.head()
```

```python
# Find the kvmer_df['key'] that are not in kvmer_ref_df['key']

outliers = kvmer_df[~kvmer_df['key'].isin(kvmer_ref_df['key'])]
len(outliers)
```

```python
# plot the histogram total_count of the outliers and kvmer_ref_df

import seaborn as sns

sns.histplot(outliers[outliers['total_count'] < 200], x='total_count', binwidth=3)
```

```python
sns.histplot(kvmer_df[kvmer_df['total_count'] < 200], x='total_count', binwidth=3)
```

```python
sns.histplot(kvmer_ref_df[kvmer_ref_df['total_count'] < 200], x='total_count', binwidth=3)
```
